#### QQ Plot (Quantitle-Qunatitle Plot) is used to find whether the given distribution is normal distribution or not

```python
import scipy.stats as stats
import pylab
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.load_dataset("iris")
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
# Let's plot the graph for every continuous feature
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
#     plt.hist(df[feature])
    plt.subplot(1,2,2)
    stats.probplot(df[feature],dist="norm",plot=pylab)
#     Generates a probability plot of sample data against the quantiles of a
# specified theoretical distribution (the normal distribution by default).
# `probplot` optionally calculates a best-fit line for the data and plots the
# results using Matplotlib or a given plot function.
    plt.show()
```

```python
plot_data(df,"sepal_width")
```

```python
# Let's plot using seaborn

def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
#     df[feature].hist()
#     plt.hist(df[feature])
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stats.probplot(df[feature],dist="norm",plot=pylab)
#     Generates a probability plot of sample data against the quantiles of a
# specified theoretical distribution (the normal distribution by default).
# `probplot` optionally calculates a best-fit line for the data and plots the
# results using Matplotlib or a given plot function.
    plt.show()
```

```python
plot_data(df,"sepal_width")
```

As per the left side figure we can conclude the distribution of data is normal distribution. The right hand side graph is plotted between predefined normal distribution quantile (stats.probplot(df[feature],**dist="norm"**,plot=pylab))(on x-axis) and df[feature] on y-axis, if all points are aligned to the red line then we can conclude that it is exact normal distribution. On this case, the points are nearly aligned with straight line, so we can conclude the distribution is normal distribution.

```python
plot_data(df,"petal_width")
```

From left side grpah we can conclude it is not a normal/ gaussian distribution , and also we can conclude the right side all points are not aligned with the straight line so it is not a normal distribution.

In the right side graph, some points are aligned with the straight line. wherever the points aligned with straight line we got a bell curve in the left side graph. But the entire distribution not followed the normal distribution.

```python
plot_data(df,"sepal_length")
```

Here the data lokks like a normal distribution as most of the points are aligned with the straight line. So we can conclude, it can be a normal distribution

```python
plot_data(df,"petal_length")
```
